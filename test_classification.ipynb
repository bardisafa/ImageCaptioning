## در قسمت پایین فرض کرده ایم که تصاویر همانند فرایند آموزش در یک فایل زیپ داده شده اند که داخل آن پوشه ای با نام خود فایل زیپ است و داخل آن تصاویر قرار گرفته اند.

```
images_arr = np.zeros([15000, TARGET_SIZE, TARGET_SIZE, 3], dtype='float32')


with ZipFile('/content/drive/My Drive/Copy of images.zip', 'r') as zipfile:
  infolist = zipfile.infolist()
  infolist = [infolist[i] for i in range(1, len(infolist)) if i%2==0]
  filename_arr = []
  for i, info in enumerate(infolist):
    filename_arr.append(info.filename[8:])     # با فرض بالا
    img = Image.open(zipfile.open(info))
    img = img.resize((TARGET_SIZE, TARGET_SIZE))
    img = img.convert('RGB')
    img = np.asarray(img).astype('float32')
    images_arr[i] = img
images_arr = images_arr[0:(i+1)]

del(infolist)
```

```
img_ids = []
for i in range(len(filename_arr)):
  x = filename_arr[i]
  y = list(df_images_info.loc[df_images_info.file_name == x,'id'])
  y = y[0]
  img_ids.append(y)
print(img_ids)
```

## لود مدل

```
model = load_model("/content/drive/My Drive/model.h5")
```

```
images_arr = preprocess_input(images_arr)
```

```
label_pred = model.predict(images_arr)
```

```
# Dictionary for finding true labels from encoded ones
# Load
read_dictionary = np.load('/content/drive/My Drive/my_file.npy',allow_pickle='TRUE').item()
print(read_dictionary) 
```

```
labels = np.zeros((len(images_arr), 1))
for i in range(len(images_arr)):
  x = int(np.argmax(label_pred[i,:]))  # Encoded label
  y = int(read_dictionary[x])
  # print(x, y)
  labels[i] = y
  labels[i] = int(labels[i][0])
```
